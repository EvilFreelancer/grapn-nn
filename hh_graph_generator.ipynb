```python
from datasets import load_dataset

vacancies = load_dataset("evilfreelancer/headhunter", split="train")
```

```python
from graph.builder import build_from_list
from mapping.relations import *
from graph.combine_graphs import *

large_graph = {}
graphs = []
for vacancy in vacancies:
    graph = build_from_list(tech_list=vacancy['technologies'], exclude_isolated=True, relations=RELATIONS)
    if len(graph['links']) == 0:
        continue
    if len(graph['nodes']) < 2:
        continue
    large_graph = combine_graphs(large_graph, graph)
```

```python
from graph.visualizer import *

visualize(large_graph, render=True)
```
